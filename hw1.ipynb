```python
from pathlib import Path
import pandas as pd
import gc
import pickle
import bz2

data_path = Path('./data/').resolve()
!ls {data_path}
```

```python
with bz2.open(data_path / '09_prepared.pkl.bz2', 'rb') as f:
    one = pickle.load(f)
one.head()
```

```python
with bz2.open(data_path / '10_prepared.pkl.bz2', 'rb') as f:
    two = pickle.load(f)
two.head()
```

```python
with bz2.open(data_path / '11_prepared.pkl.bz2', 'rb') as f:
    three = pickle.load(f)
three.head()
```

```python
df = pd.concat([one, two, three])
```

```python
df.drop(['id_card', 'id_doc'], axis=1, inplace=True)
df.head()
```

```python
strange1 = df[(df['sum'] == 0) & (df['quantity'] > 25)]
```

```python
strange1['id_card_int'].nunique()
```

```python
df = df[~df['id_card_int'].isin(strange1['id_card_int'].unique())]
df.head()
```

```python
df[(df['sum'] < 0) | (df['quantity'] < 0)]
```

```python
products = pd.read_csv(
    filepath_or_buffer='data/product_groups.txt',
    sep=';'
)
products
```

```python
df['day'] = df['date'].dt.date
df.head()
```

```python
check_sums_per_day = df.groupby(['id_card_int', 'day'])[['sum', 'quantity']].sum()
check_sums_per_day.head()
```

```python
strange2 = check_sums_per_day[check_sums_per_day['sum'] > 12500]
strange2.reset_index()['id_card_int'].nunique()
```

```python
bad_ids = set()
bad_ids |= set(strange1['id_card_int'].unique())
bad_ids |= set(strange2.reset_index()['id_card_int'].unique())
len(bad_ids)
```

```python
with bz2.open(data_path / 'bad_ids.pkl.bz2', 'wb') as f:
    pickle.dump(bad_ids, f, protocol=4)
```
